```python
import os
import paramiko
from scp import SCPClient
import logging
from datetime import datetime

logging.basicConfig(filename='backup_report.log', level=logging.INFO,
                    format='%(asctime)s - %(levelname)s - %(message)s')

SOURCE_DIRECTORY = ' '
REMOTE_SERVER = ' '
REMOTE_USERNAME = ' '
REMOTE_PASSWORD = ' '
REMOTE_DIRECTORY = ' '

def create_ssh_client(server, user, password):
    client = paramiko.SSHClient()
    client.load_system_host_keys()
    client.set_missing_host_key_policy(paramiko.WarningPolicy)
    client.connect(server, username=user, password=password)
    return client

def backup_directory():
    ssh_client = None
    try:
        ssh_client = create_ssh_client(REMOTE_SERVER, REMOTE_USERNAME, REMOTE_PASSWORD)
        scp = SCPClient(ssh_client.get_transport())

        # Copy directory to remote server
        scp.put(SOURCE_DIRECTORY, recursive=True, remote_path=REMOTE_DIRECTORY)
        scp.close()

        logging.info('Backup successful')
        print('Backup successful')
    except Exception as e:
        logging.error(f'Backup failed: {e}')
        print(f'Backup failed: {e}')
    finally:
      if ssh_client:
        ssh_client.close()

if __name__ == '__main__':
    print(f"Starting backup at {datetime.now()}")
    backup_directory()
```
